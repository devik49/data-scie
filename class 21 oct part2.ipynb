```python
import pandas as pd
```

```python
df= pd.read_csv("Automobile_data.csv")
df.head(3)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
#obj
#company ,body-style  ,engine-type,num-of-cylinders 
df['company'].nunique()
```

```python
df['company'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder

enc= LabelEncoder()
df['company ']=enc.fit_transform(df['company'])
print(df)
```

```python
df['body-style'].nunique()
```

```python
df['body-style'].unique()
```

```python
df['body-style']=df['body-style'].map({'convertible':1, 'hatchback':2, 'sedan':3, 'wagon':4, 'hardtop':5})
df
```

```python
df['engine-type'].unique()
```

```python
df['engine-type'].replace({'dohc':1, 'ohcv':2, 'ohc':3, 'l':4, 'rotor':5, 'ohcf':6},inplace=True)
df
```

```python
del df['company']
```

```python
df
```

```python
df['num-of-cylinders'].unique()
```

```python
df['num-of-cylinders'].replace({'four':4, 'six':6, 'five':5, 'three':3, 'twelve':12, 'two':2, 'eight':8},inplace=True)
df
```

```python
x=df[['body-style', 'wheel-base', 'length', 'engine-type',
       'num-of-cylinders', 'horsepower', 'average-mileage','company ']]
x
```

```python
y=df['price']
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=1)
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr.fit(xtrain,ytrain)
y_pred=lr.predict(xtest)
```

```python
from sklearn.metrics import r2_score

print("LR",r2_score(ytest,y_pred)*100)
```

```python
result=pd.DataFrame({'Actual':ytest,'Predicted':y_pred})
result
```

```python
import matplotlib.pyplot as plt
```

```python
m=list(range(1,13))

plt.plot(m,ytest,color="red",marker='^',markerfacecolor='green')
plt.scatter(m,y_pred,color='blue')
plt.show()
```

```python
from sklearn.svm import SVR

sv=SVR(gamma='auto')
sv.fit(xtrain,ytrain)

y_pred=sv.predict(xtest)
```

```python
print("LR",r2_score(ytest,y_pred)*100)
```

```python
result=pd.DataFrame({'Actual':ytest,'Predicted':y_pred})
result
```

```python
m=list(range(1,13))

plt.plot(m,ytest,color="red",marker='^',markerfacecolor='green')
plt.scatter(m,y_pred,color='blue')
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor

tree=DecisionTreeRegressor()
tree.fit(xtrain,ytrain)

y_pred=tree.predict(xtest)

print("LR",r2_score(ytest,y_pred)*100)
```

```python
result=pd.DataFrame({'Actual':ytest,'Predicted':y_pred})
result
```

```python
m=list(range(1,13))

plt.scatter(m,ytest,color="red")
plt.scatter(m,y_pred,color='blue')
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
rfc=RandomForestRegressor()
rfc.fit(xtrain,ytrain)

y_pred=rfc.predict(xtest)

print("LR",r2_score(ytest,y_pred)*100)
```

```python
result=pd.DataFrame({'Actual':ytest,'Predicted':y_pred})
result
```

```python
m=list(range(1,13))

plt.scatter(m,ytest,color="red")
plt.scatter(m,y_pred,color='blue')
plt.show()
```
