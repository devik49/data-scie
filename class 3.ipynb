```python
gas=pd.read_csv("gas_prices.csv")
print(gas.head(3))
```

```python
gas
```

```python
plt.plot(gas.Year,gas.Australia,'bH-.',label="Aus")
plt.plot(gas.Year,gas.USA,'rD-.',label="USA")
plt.plot(gas.Year,gas.UK,'gH-.',label="UK")
plt.plot(gas.Year,gas['South Korea'],'yD-.',label="SKorea")
plt.xticks(gas.Year[::2])
plt.legend(loc='upper left')
plt.show()
```

```python
countries = ['Australia','USA','UK','South Korea']

for country in countries:
    if country in gas:
        plt.plot(gas.Year,gas[country],marker='s',label=country)

plt.xticks(gas.Year[::2])
plt.legend(loc='upper left')
plt.show()
```

# pie chart


```python
lang = ['python','java','c','c++','php']
total = [25,26,15,34,45]
c=['yellow','green','pink','blue','purple']
plt.figure(figsize=(5,4),dpi=100)
e=[0,0,0,0.4,0]

plt.pie(total,labels=lang,autopct='%1.3f%%',colors=c,startangle=90,radius=1,explode = e)
plt.title("Programming language",fontdict={'name':'Times New Roman','size':30,'color':'red'})
plt.show()
```
